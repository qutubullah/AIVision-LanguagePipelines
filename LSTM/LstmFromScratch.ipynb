Stock forecasting with Lstm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

data = pd.read_csv("AMZN.csv")
data
```

```python
data = data[['Date','Close']]
data
```

```python
data['Date'] = pd.to_datetime(data["Date"])
plt.plot(data['Date'],data['Close'])
```

```python
from copy import deepcopy as dc
# n_steps=No of steps like n-1, n-2 ,n-3
def prepare_df_for_lstm(df,n_steps):
  df=dc(df)

  df.set_index('Date', inplace=True)

  for i in range(1,n_steps+1):
    df[f'Close(t-{i})'] = df['Close'].shift(i)

  df.dropna(inplace=True)

  return df
lookback = 7
shifted_df = prepare_df_for_lstm(data,lookback)
shifted_df
```

```python
shifted_df_as_np = shifted_df.to_numpy()
shifted_df_as_np
```

```python
shifted_df_as_np.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1,1))
shifted_df_as_np = scaler.fit_transform(shifted_df_as_np)

shifted_df_as_np
```

```python
X = shifted_df_as_np[:,1:]
y = shifted_df_as_np[:,0]
X.shape, y.shape
```

```python
X = dc(np.flip(X,axis=1))
X
```

```python
split_index = int(len(X) * 0.95)
split_index
```

```python
X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

The shape (batch_size, timesteps, features) is commonly used, where:

batch_size: The number of samples in each batch.
timesteps: The number of time steps in each sample or sequence.
features: The number of features at each time step.
When dealing with univariate time series data (i.e., data with only one feature, such as stock prices over time), you typically reshape the data into (batch_size, lookback, 1), where:

batch_size remains the same as the number of samples.
lookback is the number of time steps to look back in the past to predict the future.
1 indicates that there is only one feature at each time step.

```python
X_train = X_train.reshape((-1,lookback,1))
X_test = X_test.reshape((-1,lookback,1))

y_train = y_train.reshape(-1,1)
y_test = y_test.reshape((-1,1))

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train = torch.tensor(X_train).float()
X_test = torch.tensor(X_test).float()
y_train = torch.tensor(y_train).float()
y_test = torch.tensor(y_test).float()

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from torch.utils.data import Dataset
# Creating dataset for data loading
class TimeSeriesDataset(Dataset):
  def __init__(self,X,y):
    self.X = X
    self.y = y

  def __len__(self):
    return len(self.X)

  def __getitem__(self, i):
    return self.X[i], self.y[i]

train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)

train_dataset
```

```python
from torch.utils.data import DataLoader

batch_size = 16

train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size = batch_size, shuffle=False)
```

```python
for _, batch in enumerate(train_loader):
  x_batch, y_batch = batch[0],batch[1]
  print(x_batch.shape, y_batch.shape)
  break
```

```python
""" if num_stacked_layers=1, there is only one LSTM layer, and the input is processed by that single layer.
If num_stacked_layers=2, there are two LSTM layers stacked on top of each other,
and the output of the first layer is passed as input to the second layer.Increasing the number of stacked layers can potentially
improve the model's ability to capture complex relationships in the data but also increases the computational cost and the risk of overfitting,
 especially if the dataset is small. It's a hyperparameter that needs to be tuned based on the specific problem and dataset."""

#  self.hidden_size defines the dimensionality of the hidden state of the LSTM cells. It determines the number of features in the hidden state.
# hidden_size: The number of features in the hidden state of each LSTM cell.
# This parameter essentially determines the capacity of the LSTM layer to capture and represent information from the input sequence.

class LSTM(nn.Module):
  def __init__(self, input_size, hidden_size, num_stacked_layers):
    super().__init__()
    self.hidden_size = hidden_size
    self.num_stacked_layers = num_stacked_layers

    self.lstm = nn.LSTM(input_size, hidden_size, num_stacked_layers, batch_first=True)

    self.fc = nn.Linear(hidden_size,1)

  def forward(self, x):
    batch_size = x.size(0)
    h0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size)
    c0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size)

    out, _ = self.lstm(x, (h0, c0))
    out = self.fc(out[:, -1, :])
    return out

model = LSTM(1, 4, 1)
model
```

```python
def train_one_epoch():
  model.train(True)
  print(f'Epoch: {epoch + 1}')
  running_loss = 0.0

  for batch_index, batch in enumerate(train_loader):
    x_batch, y_batch = batch[0], batch[1]

    output = model(x_batch)
    loss = loss_function(output, y_batch)
    running_loss += loss.item()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if batch_index % 100 ==99:
      avg_loss_across_batches = running_loss / 100
      print('Batch {0}, Loss: {1:.3f}'.format(batch_index+1,avg_loss_across_batches))
      running_loss = 0.0
  print()
```

```python
def validate_one_epoch():
    model.train(False)
    running_loss = 0.0

    for batch_index, batch in enumerate(test_loader):
        x_batch, y_batch = batch[0], batch[1]

        with torch.no_grad():
            output = model(x_batch)
            loss = loss_function(output, y_batch)
            running_loss += loss.item()

    avg_loss_across_batches = running_loss / len(test_loader)

    print('Val Loss: {0:.3f}'.format(avg_loss_across_batches))
    print('***************************************************')
    print()
```

```python
learning_rate = 0.001
num_epochs = 10
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    train_one_epoch()
    validate_one_epoch()
```

```python
with torch.no_grad():
    predicted = model(X_train).to('cpu').numpy()

plt.plot(y_train, label='Actual Close')
plt.plot(predicted, label='Predicted Close')
plt.xlabel('Day')
plt.ylabel('Close')
plt.legend()
plt.show()
```
